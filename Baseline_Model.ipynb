# Baseline Model & Error Analysis

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

RSEED = 42
```

```python
# Read the data

df = pd.read_csv('data/the_data_we_work_with.csv')
df.head()
```

The Baseline Model will be a model that takes the Mean of the Dataset (Train) and predicts this value for any tourist

```python
y = df.total_cost
X = df.drop('total_cost', axis=1)

X = pd.get_dummies(X, drop_first=True)
X
```

```python
# Train Test Split



X_train, X_test, y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=RSEED)
```

```python
mean_value = y.mean()
print(mean_value)
```

```python
# Get an array of the same length as the test dataset
y_predicted = pd.Series(np.full_like(y_test, mean_value))
```

```python
def calc_mpe(y_true, y_predicted):
    mpe = np.mean(np.power((y_true-y_predicted)/y_true, 2))*100
    mpe = np.sqrt(mpe)
    return mpe
```

```python
# Metrics of the Baseline Model

print("RMSE: " , np.round(np.power(mean_squared_error(y_test, y_predicted),1/2), 2)*0.00036)
print("MAE: " , np.round(mean_absolute_error(y_test, y_predicted), 2)*0.00036)
print("MPE: ", calc_mpe(y_test, y_predicted))
```
